## Import and Read Dataset

```python
# 'pandas' merupakan library yang digunakan untuk analisa data
# Panggil fungsi dari 'pandas' sebagai 'pd'
import pandas as pd 

# Baca file csv menggunakan Pandas
data = pd.read_csv('IndonesiaCryptoHunter.csv')

# Tampilkan daftar kolom
print(list(data.columns))
```

## Preprossesing Data

```python
# Drop kolom yang tidak diperlukan
data.drop(['has_email', 'has_phone', 'has_hashtag', 'is_bot_command'], axis=1, inplace=True)

# Tampilkan daftar kolom
print(list(data.columns))
```

### Keterangan :


*   axis=1: parameter ini menunjukkan bahwa yang akan dihapus adalah kolom pada dataset. Nilai 1 merepresentasikan kolom sedangkan nilai 0 merepresentasikan baris. Dalam hal ini, kita ingin menghapus kolom sehingga nilai parameter adalah 1.

*   inplace=True: parameter ini menunjukkan bahwa dataset akan diubah secara langsung dan tidak akan menghasilkan dataset baru. Jika nilai parameter ini adalah True, maka dataset akan diubah secara langsung. Jika nilai parameter ini adalah False (default), maka dataset baru akan dihasilkan.



## Show Chat Count In 2022 for 12 months

```python
# Konversi kolom 'date' menjadi tipe data datetime
data['date'] = pd.to_datetime(data['date'])

# Filter data dari tanggal 2022-01-02 hingga 2022-12-31
# Format date : years-month-day
filtered_data_jan = data[(data['date'] >= '2022-01-02') & (data['date'] <= '2022-12-31')]
filtered_data_feb = data[(data['date'] >= '2022-02-01') & (data['date'] <= '2022-12-28')]
filtered_data_mar = data[(data['date'] >= '2022-03-01') & (data['date'] <= '2022-12-31')]
filtered_data_apr = data[(data['date'] >= '2022-04-01') & (data['date'] <= '2022-12-30')]
filtered_data_may = data[(data['date'] >= '2022-05-01') & (data['date'] <= '2022-12-31')]
filtered_data_jun = data[(data['date'] >= '2022-06-01') & (data['date'] <= '2022-12-30')]
filtered_data_jul = data[(data['date'] >= '2022-07-01') & (data['date'] <= '2022-12-31')]
filtered_data_aug = data[(data['date'] >= '2022-08-01') & (data['date'] <= '2022-12-31')]
filtered_data_sep = data[(data['date'] >= '2022-09-01') & (data['date'] <= '2022-12-30')]
filtered_data_oct = data[(data['date'] >= '2022-10-01') & (data['date'] <= '2022-12-31')]
filtered_data_nov = data[(data['date'] >= '2022-11-01') & (data['date'] <= '2022-12-30')]
filtered_data_dec = data[(data['date'] >= '2022-12-01') & (data['date'] <= '2022-12-31')]

# Hitung jumlah baris msg_content
jumlah_chat_jan = filtered_data_jan['msg_content'].count()
jumlah_chat_feb = filtered_data_feb['msg_content'].count()
jumlah_chat_mar = filtered_data_mar['msg_content'].count()
jumlah_chat_apr = filtered_data_apr['msg_content'].count()
jumlah_chat_may = filtered_data_may['msg_content'].count()
jumlah_chat_jun = filtered_data_jun['msg_content'].count()
jumlah_chat_jul = filtered_data_jul['msg_content'].count()
jumlah_chat_aug = filtered_data_aug['msg_content'].count()
jumlah_chat_sep = filtered_data_sep['msg_content'].count()
jumlah_chat_oct = filtered_data_oct['msg_content'].count()
jumlah_chat_nov = filtered_data_nov['msg_content'].count()
jumlah_chat_dec = filtered_data_dec['msg_content'].count()

# Tampilkan hasil
print("Chat Count")
print("Januari :", jumlah_chat_jan, "Chat")
print("Februari :", jumlah_chat_feb, "Chat")
print("Maret :", jumlah_chat_mar, "Chat")
print("April :", jumlah_chat_apr, "Chat")
print("Mei :", jumlah_chat_may, "Chat")
print("Juni :", jumlah_chat_jun, "Chat")
print("Juli :", jumlah_chat_jul, "Chat")
print("Agustus :", jumlah_chat_aug, "Chat")
print("September :", jumlah_chat_sep, "Chat")
print("Oktober :", jumlah_chat_oct, "Chat")
print("November :", jumlah_chat_nov, "Chat")
print("Desember :", jumlah_chat_dec, "Chat")
```

## Show on Line Chart

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Buat list untuk menyimpan jumlah chat
chat_count = [
    jumlah_chat_jan, jumlah_chat_feb, jumlah_chat_mar, jumlah_chat_apr,
    jumlah_chat_may, jumlah_chat_jun, jumlah_chat_jul, jumlah_chat_aug,
    jumlah_chat_sep, jumlah_chat_oct, jumlah_chat_nov, jumlah_chat_dec
]

# Buat list untuk menyimpan nama bulan
month_name = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

# Buat plot line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x=month_name, y=chat_count)
plt.title("Chat Count by Month (2022)")
plt.xlabel("Month")
plt.ylabel("Chat Count")
plt.show()
```

### Keterangan : 

*   'plt.figure()' digunakan untuk membuat sebuah figure baru, lalu parameter 'figsize' digunakan untuk mengatur ukuran dari figure yang akan dibuat.

*   Format parameter 'figsize' adalah tuple (lebar, tinggi) yang menyatakan ukuran figure dalam inch. Pada contoh di atas, ukuran figure adalah 12 inch lebar dan 6 inch tinggi.

## Show on Pie Chart

```python
import matplotlib.pyplot as plt

# Define data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
chat_counts = [jumlah_chat_jan, jumlah_chat_feb, jumlah_chat_mar, jumlah_chat_apr, jumlah_chat_may, jumlah_chat_jun,
               jumlah_chat_jul, jumlah_chat_aug, jumlah_chat_sep, jumlah_chat_oct, jumlah_chat_nov, jumlah_chat_dec]

# Create pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(chat_counts, labels=months, autopct='%1.1f%%', startangle=90)
ax.set_title('Chat Count by Month (2022)')

# Show plot
plt.show()
```

### Keterangan :   

*   'autopct' adalah singkatan dari "auto percent", yaitu  tampilan persentase pada setiap bagian dari diagram lingkaran. Argumen ini digunakan untuk menentukan format tampilan persentase pada setiap bagian dari diagram lingkaran. Nilai defaultnya adalah None, artinya tidak ada persentase yang ditampilkan. %1.1f%% menunjukkan bahwa angka persentase akan ditampilkan dengan format bilangan bulat (1 digit di depan koma) dan satu angka di belakang koma, diikuti oleh tanda persen.

*   'startangle' adalah argumen yang digunakan untuk menentukan sudut awal dari diagram lingkaran. Nilai defaultnya adalah 0, yang berarti bagian pertama dimulai dari sumbu x positif. Dalam contoh startangle=90, bagian pertama dimulai dari sumbu y positif, sehingga diagram lingkaran diputar 90 derajat searah jarum jam.
