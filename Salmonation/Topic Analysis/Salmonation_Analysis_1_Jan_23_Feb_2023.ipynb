# **Salmonation Community Topic Analysis**

```python
import pandas as pd
import nltk
from collections import Counter

# Load dataset
df = pd.read_csv('SalmonationOfficial.csv')

# drop rows with NaN values in msg_content column
df.dropna(subset=['msg_content'], inplace=True)
df.dropna(subset=['sender'], inplace=True)

# menghapus 'msg_content' dari list sender
df = df[df['sender'] != 'Rose']
df = df[df['sender'] != 'BobbyBuyBot']
df = df[df['sender'] != 'CoinMarketCap Price Bot']
df = df[df['sender'] != 'Salmonation Secretary']
df = df[df['sender'] != 'SALMONATION ANNOUNCEMENT']

# Preprocessing
df['msg_content'] = df['msg_content'].str.lower() # lowercase
#df['msg_content'] = df['msg_content'].str.replace(r'http\S+|www.\S+|t.me\S+','',regex=True) # remove urls and telegram links
#df['msg_content'] = df['msg_content'].str.replace(r'[^\w\s]','') # remove punctuation

# Split each message into pairs of words
tokens = df['msg_content'].str.split().apply(lambda x: [x[i:i+1] for i in range(0, len(x)-1, 2)])

# Flatten list of lists
tokens = [pair for sublist in tokens for pair in sublist]

# Combine pairs of words into single string
tokens = [' '.join(pair) for pair in tokens]

# Count frequency of each pair of words
counter = Counter(tokens)

# Get 10 most common pairs of words
top_pairs = counter.most_common(100)

# Print the pairs and their frequencies
for pair, freq in top_pairs:
    print(pair + ' (' + str(freq) + ')\n')

# Export output to txt file
with open('Salmonation_topWords.txt', 'w') as f:
    f.write('Top 100 words:\n')
    for pair, freq in top_pairs:
        f.write(pair + ' (' + str(freq) + ')\n')
```

# **SUI TOPIC ANALYSIS**

##Show All Messages Containing 'sui'

```python
import pandas as pd

# Filter messages containing 'sui'
df_stringKey = df[df['msg_content'].str.contains('sui', case=False)]

# Print messages
for i, msg in enumerate(df_stringKey['msg_content']):
  print(str(i+1) + '. ' + msg)
```

## Show Messages Content with Sender

```python
# Add 'sender' column to the dataframe
df_stringKey['sender_msg'] = df_stringKey['sender'] + ': ' + df_stringKey['msg_content']

# Print messages with line numbers and sender
for i, msg in enumerate(df_stringKey['sender_msg']):
    print(str(i+1) + '. ' + msg)
```

## Count Messages per Sender (contains 'sui')

```python
# Count messages containing 'sui' by sender
counts = df_stringKey['sender'].value_counts()
print(counts)

# Save output to file
with open('User_Chat_Count.txt', 'w') as f:
    f.write(counts.to_string())
```
